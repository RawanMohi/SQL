```python
###find the number of availalbe copies of Dracula      */

cursor.execute('''
SELECT 212 - COUNT(l.bookid) AS Available_Books FROM books b
LEFT JOIN loans l ON b.bookid = l.bookid 
WHERE b.title = 'Dracula' AND l.ReturnedDate IS NULL ; ''')


chec = cursor.fetchall()
chec 
```

```python
#/* current total loans of the book */* Dracula Book *

cursor.execute('''select Title , count(*) as Loans_Dracula from loans as l
inner join books AS b ON l.bookid = b.BookID
where Title  = 'Dracula' and ReturnedDate IS NOT NULL;''')
 
loan = cursor.fetchall()
loan 
```

```python
#/* Check out Books                                 */

cursor.execute ('''SELECT title ,count(*) AS Checkout_count FROM loans as l 
INNER JOIN books AS b ON b.BookID = l.BookID 
group by b.Title
order by Checkout_count ASC ;''')

books = cursor.fetchall()
books
```

```python
##* generate a report of books due back on July 13, 2020  with patron contact information                      */

cursor.execute('''select B.Title ,P.FirstName ,P.Email ,b.Barcode,l.DueDate,l.ReturnedDate from books as b 
INNER JOIN loans as l ON l.BOOKID = b.BOOKID 
INNER JOIN patrons AS p ON l.PatronID = p.PatronID
WHERE l.DueDate = '2020-07-13';''')

books = cursor.fetchall()
books
```

```python

#/* Return books to the library                         */
## All Returned books
cursor.execute('''SELECT b.Title, COUNT(l.BookID) AS CountReturnedBooks
FROM Books AS b
INNER JOIN Loans AS l ON l.BookID = b.BookID
WHERE l.ReturnedDate IS NOT NULL
GROUP BY b.Title;''')
    
books = cursor.fetchall()
books
```

```python
#/* Encourage Patrons to check out books                */
#/* generate a report of showing 10 patrons who have
#checked out the fewest books.   

cursor.execute('''select p.FirstName,count(*) as lowest_checkout from loans as l
inner join patrons as p ON p.patronid = l.patronid 
group by p.patronid ,p.FirstName
order by lowest_checkout ASC
limit 10 ;''')

books = cursor.fetchall()
books
```

```python

#/* Find books to feature for an event                  
 #create a list of books from 1890s that are
 #currently available                                    

cursor.execute('''select b.Published ,b.Title ,l.ReturnedDate from books as b 
 Inner join loans as l ON L.BookID = L.BookID 
where b.Published like '189%'  ;''')

Returnedbooks = cursor.fetchall()
Returnedbooks
```

```python

#/* Book Statistics 
#/#* create a report to show how many books were 
#published each year.                                    */

cursor.execute('''SELECT Published, COUNT(DISTINCT(Title)) AS TotalNumberOfPublishedBooks FROM Books
GROUP BY Published
ORDER BY TotalNumberOfPublishedBooks DESC;''')

year_counter_publish = cursor.fetchall()
year_counter_publish 
```

```python

#/* Book Statistics                                           */
#/* create a report to show 5 most popular Books to check out */

cursor.execute('''SELECT b.Title, count(loanid) as CheckoutCount  FROM BOOKS AS b
INNER JOIN LOANS AS l ON b.bookid = l.bookid 
group by b.Title 
order by CheckoutCount DESC
limit 5;''')
books = cursor.fetchall()
books
```
